```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def gradient_descent(x,y):
    lr=0.001
    m=0
    b=0
    n=len(x)
    plt.scatter(x,y,color='red',linewidths=5)
    count=0
    yp=0
    i=0
    while sum(abs((y-yp)))>0.2:
        yp=m*x+b
        if(i==count):
            plt.plot(x,yp,color='green')
            count+=5
        i+=1
        md=-(2/n)*sum(x*(y-yp))
        bd=-(2/n)*sum(y-yp)
        m=m-lr*md
        b=b-lr*bd
        
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
gradient_descent(x,y)
```
